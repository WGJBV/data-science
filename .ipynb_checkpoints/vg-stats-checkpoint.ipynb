```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./vgsales.csv')
```

```python
df.head()
```

```python
#Which company is the most common video game publisher?
```

```python
Publishers = pd.DataFrame(df['Publisher'].groupby(df.Publisher).count())
Publishers.sort_values('Publisher', ascending=False)
```

```python
#What’s the most common platform?
```

```python
Platforms = pd.DataFrame(df['Platform'].groupby(df.Platform).count())
Platforms.sort_values('Platform', ascending=False)
```

```python
#What about the most common genre?
```

```python
Genres = pd.DataFrame(df['Genre'].groupby(df.Genre).count())
Genres.sort_values('Genre', ascending=False)
```

```python
#What are the top 20 highest grossing games?
```

```python
Gross = pd.DataFrame(df[['Global_Sales', 'Name']])
Gross.head(20)
```

```python
#For North American video game sales, what’s the median? Provide a secondary
#output showing 'about' ten games surrounding the median sales output
```

```python
North_Sales = pd.DataFrame(df['NA_Sales'])
North_Sales.median()
```

```python
df[(df['NA_Sales'] == 0.08)].head(10)
```

```python
#For the top-selling game of all time, how many standard deviations above/below the mean are its sales for North America?
```

```python
nastd = pd.DataFrame(df['NA_Sales'])
nastd.std()
```

```python
#The Nintendo Wii seems to have outdone itself with games. How does its average number of sales compare with all of the
#other platforms?
```

```python
grosswii = pd.DataFrame(df[['Platform', 'Global_Sales']]).groupby(df.Platform).mean()
grosswii.sort_values('Global_Sales', ascending=False)
```

```python
#What were the top grossing games from 1990-2000.
```

```python
grossteen = pd.DataFrame(df[(df['Year'] >= 1990.0) & (df['Year'] <= 2000.0)][['Global_Sales', 'Name']]).groupby(df.Name)
grossteen.head(5)
```

```python
#What Company has the most Simulation games out there?
```

```python
simulationco = pd.DataFrame(df[df['Genre'] == 'Simulation']['Publisher']).groupby(df.Publisher).count()
simulationco.sort_values('Publisher', ascending=False)
```

```python
#Globally, what genres are the highest grossing on average?
```

```python
genregross = pd.DataFrame(df[['Genre', 'Global_Sales']]).groupby(df.Genre).mean()
genregross.sort_values('Global_Sales', ascending=False)
```
